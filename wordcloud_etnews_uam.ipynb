```python
# txt 파일 열어주기
f = open('20230724_etnews_uam.txt', 'r')
data = f.read()
print(data)
```

```python
from konlpy.tag import Okt
from collections import Counter

# Okt 함수를 이용해 형태소 분석
okt = Okt()
line =[]

line = okt.pos(data)

n_adj =[]
# 명사 또는 형용사인 단어만 n_adj에 넣어주기
for word, tag in line:
    if tag in ['Noun','Adjective']:
        n_adj.append(word)
# 명사 또는 형용사인 단어 및 2글자 이상인 단어 선택 시
# n_adj = [word for word, tag in line if tag in ['Noun','Adjective'] and len(word) > 1]
print(n_adj)
```

```python
#제외할 단어 추가
stop_words = "발행 일 사상 최대 기술 사업 업계 최고 를 중 및 입 달 점 국내 계획 수 또한 분 은 태 등 고 를 은 는 이 가 듯 계 입 연 강 것 위 있는 있다 개발 위해 추진 기자 통해 말 다양한 개 올해 사장 과 새로운 있도록 의 용 쇼 초 제 특 지난해 며 적극 전 로 비 면서 시 창 또 더 후 도 그 내 총 " #추가할 때 띄어쓰기로 추가해주기
stop_words = set(stop_words.split(' '))
# 불용어를 제외한 단어만 남기기
n_adj = [word for word in n_adj if not word in stop_words]
```

```python
#가장 많이 나온 단어 200개 저장
counts = Counter(n_adj)
tags = counts.most_common(200)
```

```python
from wordcloud import WordCloud

###WordCloud(워드크라우드) 만들기###
#폰트지정
font='NanumGothic.ttf'
word_cloud = WordCloud(font_path=font, background_color='black', width = 400, height = 400, max_font_size=80, colormap='Accent').generate_from_frequencies(dict(tags))

# 사이즈 설정 및 화면에 출력
import matplotlib.pyplot as plt
plt.figure(figsize=(16,16))
plt.imshow(word_cloud)
plt.axis('off')
#워드클라우드 사진으로 저장
plt.savefig('wc_uam.png', bbox_inches='tight')
plt.show()
```
